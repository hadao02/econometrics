```python
import pandas as pd

# Read data from CSV file
df = pd.read_csv('D:/Personal project/IMDB/econometrics/index_merged_data.csv')

# Print column names
print('Column names:', df.columns)

# Print data types of each column
print('Data types:', df.dtypes)
```

```python
df.head(15)
```

```python
columns = ['PropertyRights', 'JudicialEffectiveness', 'GovernmentIntegrity', 'TaxBurden', 'GovSpending', 'FiscalHealth', 'BusinessFreedom', 'LaborFreedom', 'MonetaryFreedom', 'TradeFreedom', 'InvestmentFreedom', 'FinancialFreedom', 'TariffRate', 'IncomeTaxRate', 'CorporateTaxRate', 'TaxBurdenRateOfGDP', 'GovExpenditureRateOfGDP', 'PopulationMillions', 'UnemploymentRate', 'InflationRate', 'FDIInflowMillions', 'PublicDebtRateOfGDP']

grouped = df.groupby('CountryName')
result = grouped[columns].pct_change()
result[['CountryName','Year','Region','GDPGrowthRate']] = df[['CountryName','year','Region','GDPGrowthRate']]
```

```python
result
```

```python
cols = ['CountryName', 'Region', 'Year','GDPGrowthRate'] + [col for col in result.columns if col not in ['CountryName', 'Region', 'Year','GDPGrowthRate']]

# Select columns from result DataFrame in desired order
result = result[cols]
result.head(15)
```

```python
result.to_csv('D:/Personal project/IMDB/econometrics/pct_change.csv', index=False)
```
